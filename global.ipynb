```python
jobs_start = []
for level in ['Internship', 'Entry level']:
    jobs_start += jobs.jobs_by_level[level]

jobs_end = []
for level in ['Associate', 'Mid-Senior', 'Senior', 'Director', 'Executive']:
    jobs_end += jobs.jobs_by_level[level]

start_p = jobs.get_skill_percents(jobs_start)
end_p = jobs.get_skill_percents(jobs_end)


pairs = []
for skill in jobs.skills:
    delta = start_p[skill] - end_p[skill]
    if abs(delta) >= 2:
        pairs.append((delta, skill))

pairs.sort()

"""
for p, s in pairs:
    if abs(p) >= 2:
        print(s, p)
"""
import pandas as pd      
df = pd.DataFrame({'skill': [s for _, s in pairs], 'percent': [p for p, _ in pairs]})
ax = df.plot.bar(x='skill', y='percent', rot=90, figsize=(30,10))
for item in ([ax.title, ax.xaxis.label, ax.yaxis.label] +
             ax.get_xticklabels() + ax.get_yticklabels()):
    item.set_fontsize(20)
ax
```

```python
programming_languages = ['java', 'python', 'r', 'scala', 'c', 'c++', 'c#', 'sql', 'haskell', 'javascript']
p = []
for lang in programming_languages:
    p.append(jobs.skill_percents[lang])
df = pd.DataFrame({'language': programming_languages, 'percent': p})
ax = df.plot.bar(x='language', y='percent', rot=90, figsize=(20,5))
for item in ([ax.title, ax.xaxis.label, ax.yaxis.label] +
             ax.get_xticklabels() + ax.get_yticklabels()):
    item.set_fontsize(20)
ax
```
